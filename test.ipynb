```python
import sys, subprocess

def ensure(pkg):
    subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])

try:
    import pandas_datareader
except ImportError:
    ensure("pandas-datareader==0.10.0")
    import pandas_datareader
```

```python
import pandas as pd
import pandas_datareader.data as web
from datetime import datetime, timedelta

# Stooq works best with explicit dates
end = datetime.utcnow().date()
start = end - timedelta(days=365)

df = web.DataReader("BTCUSD", "stooq", start=start, end=end)
print(df)
# Stooq returns newest-first; sort ascending like yfinance
df = df.sort_index()

# Match yfinance-style output as closely as possible
df["Adj Close"] = df["Close"]
df.index = pd.to_datetime(df.index).tz_localize(None).normalize()

print(df.shape)
print(df.tail())
```
